### **Google Search Analysis with Python**
Google doesn’t give much access to the data about daily search queries, but another application of google known as Google Trends can be used for the task of Google search analysis. Google Trends provides an API that can be used to analyze the daily searches on Google. This API is known as pytrends, you can easily install it in your systems by using the pip command; pip install pytrends.

```python
from IPython.core.display import publish_display_data
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import cv2

from pytrends.request import TrendReq
trends = TrendReq()

%matplotlib inline
```

Here I will be analyzing the Google search trends on the queries based on “**Data Science**”, so let’s create a DataFrame of the top 30 countries which search for “**Data Science**” on Google

```python
trends.build_payload(kw_list=["Data Science"])
data = trends.interest_by_region()
data = data.sort_values(by="Data Science", ascending=False)
data = data.head(30)
print(data)
```

So, according to the above results, the search queries based on “Data Science” are mostly done in Singapore & India. We can also visualize this data using a bar chart




```python
data.reset_index().plot(x="geoName", y="Data Science",
                        figsize=(20,15), kind="bar")
plt.style.use('fivethirtyeight')
plt.rc('font', size=16)
plt.title('Total Google Searches for Data Science', fontweight='bold')
plt.show()
```

So as we all know that **Data Science** has been the focus of so many countries and students for the last 3-4 years, so let’s have a look at the trend of searches to see how the total search queries based on “Data Science” increased or decreased on Google

```python
data = TrendReq(hl='en-US', tz=360)
data.build_payload(kw_list=['Data Science'])
data = data.interest_over_time()
fig, ax = plt.subplots(figsize=(20, 15))
data['Data Science'].plot()
plt.style.use('fivethirtyeight')
plt.title('Total Google Searches for Data Science', fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Total Count')
plt.show()
```

### **Conclusion**
So we can see that searches based on “Data Science” on Google started to increase in 2018 and the highest searches were done in 2021-2022 . This is how we can analyze Google searches based on any keyword. A business can perform Google search analysis to understand what people are looking for on Google at any given time.
